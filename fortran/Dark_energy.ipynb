```python
import numpy as np
import matplotlib.pyplot as plt

file_path1 = 'planck_2018_lenspotentialCls.dat'
file_path2 = 'planck_2018_lenspotentialCls_Dark1.dat'

with open(file_path1, 'r') as file:
    content1 = file.read()
with open(file_path2, 'r') as file:
    content2 = file.read()
    
liste1 = list(content1.split())
liste2 = list(content2.split())

l = []
maxl = 3500

del liste1[0]
del liste1[0]
del liste1[0]
del liste1[0]
del liste1[0]
del liste1[0]
del liste1[0]
del liste1[0]
del liste1[0]

del liste2[0]
del liste2[0]
del liste2[0]
del liste2[0]
del liste2[0]
del liste2[0]
del liste2[0]
del liste2[0]
del liste2[0]

for i in range(maxl-2):
    l.append(float(liste1[i*8]))

Dltt1 = []
Dltt2 = []

del liste1[0]
del liste2[0]

for i in range(maxl-2):
    Dltt1.append(float(liste1[i*8]))

for i in range(maxl-2):
    Dltt2.append(float(liste2[i*8]))

relativ_error = []

for i in range(maxl-2):
    relativ_error.append((Dltt2[i]-Dltt1[i])/Dltt1[i])

plt.plot(l,Dltt1, 'k')
plt.plot(l,Dltt2 ,'r--')
plt.ylabel('Dltt (µK²)')
plt.xlabel('echelle angulaire l')
plt.legend(['λCDM', 'EDE'])
plt.savefig('Power Spectrums')
```
