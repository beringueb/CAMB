```python
import numpy as np
import matplotlib.pyplot as plt

file_path1 = 'test_rayleigh_lensedCls.dat_6_6'

with open(file_path1, 'r') as file:
    content1 = file.read()

liste1 = list(content1.split())


del liste1[0]

Dltt1 = []


for i in range(int((len(liste1)+1)/5)):
    Dltt1.append(float(liste1[i*5]))

l = np.arange(2,len(Dltt1)+2)

#print(Dltt1)

print(Dltt1[0])
print(Dltt1[1])
#pprint(Dltt1[2])
#pprint(Dltt1[3])
#pprint(Dltt1[4])
#pprint(Dltt1[5])
plt.plot(l,Dltt1)
plt.ylabel('CTT freq 6_6')
plt.xlabel('echelle angulaire (l)')
plt.savefig("CTT freq 6_6.png")
```
