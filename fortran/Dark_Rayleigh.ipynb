```python
import numpy as np
import matplotlib.pyplot as plt

#file_path1 = 'planck_2018_lensedCls_Dark.dat_4_4'
file_path1 = 'planck_2018_lensedCls.dat_4_4'
file_path2 = 'planck_2018_lensedCls.dat_1_1'

with open(file_path1, 'r') as file:
    content1 = file.read()
with open(file_path2, 'r') as file:
    content2 = file.read()
    
liste1 = list(content1.split())
liste2 = list(content2.split())

l = []
maxl = 4500

#del liste11[0]
#del liste22[0]

print(liste1[0:10])

for i in range(maxl-2):
    l.append(float(liste1[i*5]))

Dltt1 = []
Dltt2 = []

del liste1[0]
del liste2[0]

for i in range(maxl-2):
    Dltt1.append(float(liste1[i*5]))

for i in range(maxl-2):
    Dltt2.append(float(liste2[i*5]))

relativ_error = []

for i in range(maxl-2):
    relativ_error.append((Dltt1[i]-Dltt2[i])/Dltt2[i])

print(Dltt1[0:10])
print(Dltt1[0])

plt.plot(l,Dltt1, 'k')
plt.plot(l,Dltt2 ,'r--')
plt.ylabel('Dltt (µK²)')
plt.xlabel('echelle angulaire l')
#plt.legend(['λCDM', 'EDE'])
plt.legend(['EDE_600_600', 'λCDM_600_600'])
plt.savefig('Rayleigh Power Spectrums comparaison')
```

```python
plt.plot(l,relativ_error)
plt.ylim([-0.07,0.005])
plt.ylabel('(µK)²')
plt.xlabel('echelle angulaire l')
plt.title('Fraction rationelle "(EDE-λCDM/λCDM)"')
plt.savefig('Rayleigh Fraction rationelle comparaison')
```
