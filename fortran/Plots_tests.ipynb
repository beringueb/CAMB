```python
import numpy as np
import matplotlib.pyplot as plt

file_path1 = 'planck_2018_lensedCls.dat_6_6'

with open(file_path1, 'r') as file:
    content1 = file.read()

liste1 = list(content1.split())


del liste1[0]

Dltt1 = []


for i in range(int((len(liste1)+1)/5)):
    Dltt1.append(float(liste1[i*5]))

l = np.arange(2,len(Dltt1)+2)

print(Dltt1[0])
print(Dltt1[10])
print(Dltt1[100])
print(Dltt1[1000])
plt.plot(l,Dltt1)
plt.ylabel('CTT_6_6')
plt.xlabel('echelle angulaire (l)')
#plt.yscale('log')
plt.ylim((0,10))
plt.savefig("CTT_6_6.png")
```
